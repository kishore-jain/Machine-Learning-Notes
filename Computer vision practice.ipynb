```python
#image translation
import cv2
import numpy as np 
import imutils
jeep = cv2.imread(r"C:\Users\SASIKUMAR\Desktop\rk\img\wrangler.jpg")

# create translation matrix
TM = np.float32([[1,0,60],[0,1,60]])
shift = cv2.warpAffine(jeep,TM,(jeep.shape[1],jeep.shape[0]))# 1 - width 0-height
cv2.imshow('shfited down and right',shift)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
def translate(jeep,shiftX,shiftY):
    TM = np.float32([[1,0,shiftX],[0,1,shiftY]])
    shift = cv2.warpAffine(jeep,TM,(jeep.shape[1],jeep.shape[0]))
    return shift
```

```python
# detecting Faces
#-----------------

# for detecting faces convert a color image into gray-scale 
# also download and paste haar-cascade frontal face classifier xml file in working space 
import cv2
face_cascade = cv2.CascadeClassifier(r"C:\Python38\haarcascade_frontalface_default.xml")
img = cv2.imread(r"C:\Users\SASIKUMAR\Desktop\rk\img\crowd.jpg")
gray_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
faces = face_cascade.detectMultiScale(gray_img,1.1,4)
for (x,y,w,h) in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)
cv2.imshow('detected faces',img)
cv2.waitKey()
```

```python
# detecting faces in a video
import cv2
face_cascade = cv2.CascadeClassifier(r"C:\Python38\haarcascade_frontalface_default.xml")
v_cap = cv2.VideoCapture(0)
#v_cap = cv2.VideoCapture('')
while True:
    _,img = v_cap.read()
    gray_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray_img,1.1,4)
    for (x,y,w,h) in faces:
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)
    cv2.imshow('detected faces',img)
    p=cv2.waitKey(50)
    if p==45:
        break
v_cap.release()
```

```python
#image segmentation(kmeans)
#process of partitioning the image into multiple different regions
#goal is to change the representation of image into easy and meaningful image
import cv2
import matplotlib.pyplot as plt
import numpy as np
image=cv2.imread(r"C:\Users\SASIKUMAR\Desktop\rk\img\img_py3.jpg")
image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
image=cv2.imread(r"C:\Users\SASIKUMAR\Desktop\rk\img\img_py3.jpg")
image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
image.shape
```

```python
pixel_values=image.reshape(-1,3)
pixel_values=np.float32(pixel_values)
```

```python
pixel_values.shape
```

```python
#stopping criteria
criteria=(cv2.TERM_CRITERIA_EPS+cv2.TERM_CRITERIA_MAX_ITER,100,0.5)
```

```python
k=3
_,labels,centers=cv2.kmeans(pixel_values,k,None,criteria,10,cv2.KMEANS_RANDOM_CENTERS)
centers=np.uint8(centers)
labels=labels.flatten()
```

```python
#construct the segmented image
seg_image=centers[labels.flatten()]
```

```python
#converting the image into original
seg_image=seg_image.reshape(image.shape)
plt.imshow(seg_image)
```
