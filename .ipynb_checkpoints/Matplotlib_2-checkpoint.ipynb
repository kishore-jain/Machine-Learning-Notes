```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
# pie chart

count = [26,51,31,13,42]
color_code = ['m','c','y','r','g']
district = ['sivagangai','madurai','theni','dindigul','trichy']
plt.pie(count,colors=color_code,labels=district,explode=[0,0,0.2,0,0])
plt.title('Corona count')
plt.show()
```

```python
# scatter plot
from pylab import randn

a = randn(5)
b = randn(5)

c = randn(10)
d = randn(10)

plt.scatter(a,b,color='r')

plt.scatter(c,d,color='b')

plt.legend(['a-b','c-d'])

plt.show()
```

```python
# box and whisker plot
df = pd.DataFrame(np.random.rand(10,5),columns = ['a','b','c','d','e'])
#df
df.plot.box()
# outlier-the different between actual and expectution
```

```python
# heatmap
value = [{1,3,5,9,6},{7,4,5,2,7},{7,3,4,2,8},{4,7,3,2,1},{3,6,2,9,5}]
row = ['r1','r2','r3','r4','r5']
col = ['c1','c2','c3','c4','c5']
df1 = pd.DataFrame(value,index=row,columns=col)
plt.pcolor(df1)
plt.show()
```

```python
# bubble chart
a = np.random.rand(50)
b = np.random.rand(50)
c = np.random.rand(50)

colors = np.random.rand(50)
# s = size and c= colour
plt.scatter(a,b,s=c*400,c=colors)
plt.show()
```

```python
#3D chart

import matplotlib.pyplot as plt 

from mpl_toolkits.mplot3d import axes3d
# rc default is in matplotlib 2d frame(r-rows,c-columns)
chart_style = plt.figure()# mentions the machine to change from 2d to 3d
chart3d = chart_style.add_subplot(111,projection='3d')
# interpolation
x,y,z = axes3d.get_test_data(0.15)
chart3d.plot_wireframe(x,y,z,color = 'g',rstride=1,cstride=1)
#print(x)
#print(y)
#print(z)
plt.show()
```
